```python
batsman_stats = df5.groupby(['Match_Id','Striker_Id']).agg({
    'Batsman_Scored':'sum'
}).reset_index().rename(columns={'Striker_Id':'Player_Id', 'Batsman_Scored':'Runs'})

```

```python
# Filter rows where a dismissal happened
wickets_df = df5[df5['Player_dissimal_Id'].notna()]

bowler_stats = wickets_df.groupby(['Match_Id','Bowler_Id']).agg({
    'Player_dissimal_Id':'count'   # Number of wickets
}).reset_index().rename(columns={'Bowler_Id':'Player_Id','Player_dissimal_Id':'Wickets'})
```

```python
# Merge runs
master_df = master_df.merge(batsman_stats, on=['Match_Id','Player_Id'], how='left')

# Merge wickets
master_df = master_df.merge(bowler_stats, on=['Match_Id','Player_Id'], how='left')

# Fill NaNs with 0
master_df['Runs'] = master_df['Runs'].fillna(0)
master_df['Wickets'] = master_df['Wickets'].fillna(0)
```

```python
master_df['Runs'] = pd.to_numeric(master_df['Runs'], errors='coerce').fillna(0)
master_df['Wickets'] = pd.to_numeric(master_df['Wickets'], errors='coerce').fillna(0)
```

```python
master_df.columns = master_df.columns.str.strip()
bat_hand_counts = master_df['Batting_Hand'].value_counts().reset_index()
bat_hand_counts.columns = ['Batting_Hand', 'Count']  # rename properly
```

# AGGREGATIONS PERFORMED

## Player-level: Matches, Runs, Wickets, Avg Runs	Top players
## Team-level: Wins, Total Runs	Strongest teams
## Season-level: Matches, Runs, Wickets	Trend analysis
## Venue-level: Avg runs/match	Ground insights
## Award-level: Orange/Purple Cap counts	Star player insights
## Batting/Bowling type counts	Player composition

```python
player_stats = master_df.groupby(['Player_Id','Player_Name']).agg({
    'Match_Id':'nunique',  # Matches played
    'Runs':'sum',           # Total runs
    'Wickets':'sum'         # Total wickets
}).rename(columns={'Match_Id':'Matches_Played'}).reset_index()

print(player_stats.head())
```

```python
# Matches played, total runs, total wickets, average runs per match
player_stats = master_df.groupby(['Player_Id','Player_Name']).agg({
    'Match_Id':'nunique',    # Total matches played
    'Runs':'sum',             # Total runs scored
    'Wickets':'sum',          # Total wickets taken
}).rename(columns={'Match_Id':'Matches_Played'}).reset_index()

# Add derived metric: average runs per match
player_stats['Avg_Runs_Per_Match'] = player_stats['Runs'] / player_stats['Matches_Played']

player_stats.sort_values('Runs', ascending=False).head()
```

```python
# Team-wise total wins
team_wins = master_df.groupby('Team_Name_Winner')['Match_Id'].nunique().reset_index()
team_wins = team_wins.rename(columns={'Match_Id':'Wins'}).sort_values('Wins', ascending=False)

# Total runs scored by each team
team_runs = master_df.groupby('Team_Name_x')['Runs'].sum().reset_index().rename(columns={'Runs':'Total_Runs'})
```

```python
# Total matches, total runs, total wickets per season
season_stats = master_df.groupby('Season_Year').agg({
    'Match_Id':'nunique',
    'Runs':'sum',
    'Wickets':'sum'
}).rename(columns={'Match_Id':'Matches'}).reset_index()
```

```python
venue_stats = master_df.groupby('Venue_Name').agg({
    'Runs':'sum',
    'Wickets':'sum',
    'Match_Id':'nunique'
}).rename(columns={'Match_Id':'Matches'}).reset_index()

# Average runs per match in each venue
venue_stats['Avg_Runs_Per_Match'] = venue_stats['Runs'] / venue_stats['Matches']
```

```python
# Count how many times each player won Orange/ Purple / Man of the Series
award_stats = master_df[['Orange_Cap_Id','Purple_Cap_Id','Man_of_the_Series_Id']].apply(pd.Series.value_counts).fillna(0)
```

```python
# Distribution of batting hands
bat_hand_counts = master_df['Batting_Hand'].value_counts().reset_index().rename(columns={'index':'Batting_Hand','Batting_Hand':'Count'})

# Distribution of bowling skills
bowling_skill_counts = master_df['Bowling_Skill'].value_counts().reset_index().rename(columns={'index':'Bowling_Skill','Bowling_Skill':'Count'})
```

# VISUALIZATION BEGINS

```python
# Top 10 run scorers
top_runs = player_stats.sort_values('Runs', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='Runs', y='Player_Name', data=top_runs, hue='Player_Name', dodge=False, palette='viridis', legend=False)
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Total Runs')
plt.ylabel('Player Name')
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Season_Year', y='Runs', data=season_stats, marker='o')
plt.title('Total Runs Scored per Season')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python


# Horizontal bar chart: Team Wins
team_wins_sorted = team_wins.sort_values('Wins', ascending=True)

plt.figure(figsize=(10,6))
sns.barplot(
    x='Wins', 
    y='Team_Name_Winner', 
    data=team_wins_sorted, 
    hue='Team_Name_Winner',  # assign y to hue
    dodge=False,
    palette='coolwarm', 
    legend=False             # remove extra legend
)
plt.title('Total Wins by Teams')
plt.xlabel('Wins')
plt.ylabel('T')

```

```python
# Count of each batting hand
bat_hand_counts = master_df['Batting_Hand'].value_counts().reset_index()
bat_hand_counts.columns = ['Batting_Hand', 'Count']
# Set index and plot
batting_counts = bat_hand_counts.set_index('Batting_Hand')
batting_counts.plot(
    kind='pie',
    y='Count',
    autopct='%1.1f%%',
    figsize=(6,6),
    legend=False
)
plt.title('Batting Hand Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(player_stats[['Runs','Wickets']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation: Runs vs Wickets')
plt.show()
```
